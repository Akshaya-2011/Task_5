```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
df = pd.read_csv("/content/titanic.csv")
```

```python
print("Dataset Shape:", df.shape)
print("\n--- Info ---")
print(df.info())
```

```python
print("\n--- Summary Statistics ---")
print(df.describe(include="all"))
```

```python
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
print("\n--- Value Counts: Survived ---")
print(df['Survived'].value_counts())
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
sns.pairplot(df[numeric_cols], diag_kind="hist")
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
df[numeric_cols].hist(figsize=(12, 8), bins=20, color='skyblue')
plt.suptitle("Histograms of Numeric Columns", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='coolwarm')
plt.title("Fare vs Age Colored by Survival")
plt.show()
```
